```python
from datasets import Dataset

# Your custom data examples (you can change these sentences later)
data = {
    "text": [
        "Once upon a time, a dragon lived in a cave.",
        "The stars shone brightly in the sky.",
        "Artificial intelligence is changing the world.",
        "The little robot wanted to be a hero.",
        "In the forest, something magical was about to happen.",
        "She opened the book and got lost in a world of adventure."
    ]
}

# Convert to Hugging Face Dataset format
dataset = Dataset.from_dict(data)

# Split into training and testing (90% training, 10% testing)
dataset = dataset.train_test_split(test_size=0.1)

# Show an example
print(dataset)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
def tokenize_function(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=128)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
# Add padding token to GPT-2 tokenizer
tokenizer.pad_token = tokenizer.eos_token
model.resize_token_embeddings(len(tokenizer))
```

```python
def tokenize_function(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=128)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import DataCollatorForLanguageModeling

# Setup data collator (for language modeling)
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # GPT-2 does NOT use masked language modeling
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",           # Where to save model
    per_device_train_batch_size=2,           # Batch size for training
    per_device_eval_batch_size=2,            # Batch size for evaluation
    num_train_epochs=3,                      # Number of training epochs
    evaluation_strategy="epoch",             # Evaluate at end of each epoch
    save_strategy="epoch",                   # Save model after each epoch
    logging_dir="./logs",                    # Log directory
    logging_steps=10,                        # Log every 10 steps
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",           # Where to save model
    per_device_train_batch_size=2,           # Batch size for training
    per_device_eval_batch_size=2,            # Batch size for evaluation
    num_train_epochs=3,                      # Number of training epochs
    logging_dir="./logs",                    # Log directory
    logging_steps=10,                        # Log every 10 steps
    save_steps=500,                          # Save every 500 steps
    eval_steps=500,                          # Evaluate every 500 steps
    do_eval=True,                            # Turn on evaluation
    report_to="none"                         # ✅ Disable all logging tools (no wandb)
)

```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)

# Start training
trainer.train()
```
